# Type Token Ratio (TTR)

```python
import re
f=open ("ts.txt")
count_freq=dict()
token=0
for word in re.split("[\n.*, \-!?:_;() ' \“ \”]+",f.read()):    
    word=word.lower()
    if word in count_freq:
        count_freq[word]+=1
    else:
        count_freq[word]=1
```

```python
for key,value in count_freq.items():
    token+=value
print ("Type: ",len(count_freq))
print("Token: ",token)
print("TTR: ",len(count_freq)/token)
```

# Zipf's Law

```python
import matplotlib.pyplot as plt
y=list() 
i=1
x = sorted(count_freq.values(),reverse=True)[0:15]
for i in range(15):
    y.append(i)
 
plt.plot(x, y)
 
plt.xlabel('frequency')
plt.ylabel('rank')
 
plt.title('Zipf law: frequency inversely proportional to rank')
 
plt.show()
```

```python
import matplotlib.pyplot as plt
sortedDict=sorted(count_freq.items(),key=lambda t:t[1],reverse=True)[0:150]
x = sorted(count_freq.values(),reverse=True)[0:150]
li=list()

for key,value in sortedDict:
      li.append(len(key))
      
plt.plot(x, li)
 
plt.xlabel('frequency')
plt.ylabel('length')
 
plt.title('Zipf law: frequency inversely proportional to length')
 
plt.show()
```

# Heap's Law

```python
|V| = KN^B

Taking ln on both the sides,

ln|V| = ln(K) + Bln(N)

This is of the form of straight line i.e. y= mx + c
where, m= slope
        = B
       c= y-intercept
        = ln(K)
We can find the value of K by taking antilog of the y-intercept in the graph plotted between ln|V| & ln(N)
```

```python
import math
from numpy import *
from scipy.interpolate import *
from matplotlib.pyplot import *
f=open ("ts.txt")
count_freq=dict()
v=list()
n=list()
count=0
for word in re.split("[\n.*, \-!?:_;() ' \“ \”]+",f.read()):    
    word=word.lower()
    count+=1
    if word in count_freq:
        count_freq[word]+=1
    else:
        count_freq[word]=1
    if count%1000 == 0:
       token=0
       for key,value in count_freq.items():
          token+=value
       v.append(math.log(len(count_freq)))
       n.append(math.log(token))        
        
p1= polyfit(n,v,1)
plot(n,v)
xlabel('ln(N)')
ylabel('ln(|v|)')
show()

print("[","slope,intercept","]::",p1)
```
